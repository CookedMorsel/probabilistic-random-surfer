```python
import logging
from prob_surfer.parsing.html import get_url_data_as_html

logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    handlers=[
        # logging.FileHandler("app.log"),  # Log messages to a file
        logging.StreamHandler(),  # Log messages to the console
    ],
)
logger = logging.getLogger(__name__)

url = "https://en.wikipedia.org/wiki/Random_surfing_model"


html_data = get_url_data_as_html(url)
# with open("out.html", "w", encoding="utf-8") as f:
#     f.write(html_data)
```

```python
from lxml import html, etree



data_tree = html.fromstring(html_data)

# Convert to a pretty-printed XML string
xml_content = etree.tostring(data_tree, pretty_print=True, encoding='unicode')

# print(xml_content)
```

```python
with open('out.xml', 'wb', encoding='utf-8') as f:
    f.write(xml_content)
```
